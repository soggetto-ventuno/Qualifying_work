```python
import pandas as pd

# загрузка данных
file_path = './data/data_raw_english_columns.csv'
df = pd.read_csv(file_path, low_memory=False)

# вывод 5 первых строк
df.head(5)

```

```python

# общая информация о данных
df_description = df.describe(include='all')
df_description
```

```python
# общая информация о данных
df.info()
```

```python

# общая информация о данных
print("\nОбщая информация о данных:")
print(f"Количество строк: {len(df)}")
print(f"Количество столбцов: {len(df.columns)}")

# наличие дубликатов
print(f"\nКоличество дубликатов: {df.duplicated().sum()}")

# наличие пропущенных значений
print("\nКоличество пропущенных значений в каждом столбце:")
print(df.isna().sum())

```

```python
# получение списка столбцов с пропущенными значениями
df_missing = df.isnull().sum()

print(df_missing)
 # вывод списка столбцов с пропущенными значениями > 0 с пропущенными значениями
df_missing = df_missing[df_missing > 0]
print(df_missing)
```

```python
# Преобразование типов данных
cols_to_convert = ['cost_of_education_rub', 'avg_ege_budget', 'gdp_deflator']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# проверка результата данных
converted_info = df[cols_to_convert].info()
new_missing = df[cols_to_convert].isnull().sum()
print(df[['cost_of_education_rub', 'avg_ege_budget', 'gdp_deflator']].describe())
converted_info, new_missing[new_missing > 0]  # Show only columns with missing values after conversion
```

```python

# заполнение пропущенных значений медианой по каждому столбцу
for col in ['cost_of_education_rub', 'avg_ege_budget', 'gdp_deflator']:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
print(df[['cost_of_education_rub', 'avg_ege_budget', 'gdp_deflator']].describe())



```

```python

# остались ли пропущенные значения в этих столбцах
remaining_nans = df[['cost_of_education_rub', 'avg_ege_budget', 'gdp_deflator']].isnull().sum()
print(remaining_nans)
```

```python
df_description = df.describe(include='all')
df_description
```

```python
print(df['ege_excludes_individual'])

# преобразование значений "Да" и "Нет" в булевы значения True и False
df['ege_excludes_individual'] = df['ege_excludes_individual'].map({'Да': True, 'Нет': False})

# какие уникальные значения теперь содержит этот столбец
unique_values = df['ege_excludes_individual'].unique()



```

```python
# получение количества пропущенных значений
missing_values = df.isna().sum()

# вывод количества пропущенных значений
print("Number of missing values in each column:")
print(missing_values)

# показать только столбцы с пропущенными значениями
missing_values = missing_values[missing_values > 0]
if not missing_values.empty:
    print("\nColumns with missing values:")
    print(missing_values)
else:
    print("\nNo columns have missing values.")
```

```python
# включение опции "future.no_silent_downcasting" чтобы убрать предупреждение
pd.set_option('future.no_silent_downcasting', True)
# обработка оставшихся пропусков
numeric_fill_median = ['year', 'avg_ege_paid', 'enrolled_paid', 'enrolled_budget', 'gdp_per_capita_usd']
# заполнение пропущенных значений медианой
for col in numeric_fill_median:
    df[col] = df[col].fillna(df[col].median())



```

```python
# вставляем самое частое значение в пропуски
categorical_fill_mode = ['program_group', 'university']
for col in categorical_fill_mode:
    df[col] = df[col].fillna(df[col].mode()[0])

# вставляем False в пропуски
df['ege_excludes_individual'] = df['ege_excludes_individual'].fillna(False)
df['ege_excludes_individual'] = df['ege_excludes_individual'].infer_objects(copy=False)

# вставляем 0 в пропуски
df['ege_change_prev_year'] = df['ege_change_prev_year'].fillna(0)

# проверка оставшихся пропущенных значений
missing_values = df.isnull().sum()
print(missing_values)
print(df['ege_change_prev_year'])
print(df.head(10))
```

```python
print(df['enrolled_budget'])
df.fillna({
    'ege_change_prev_year': 0,
    'avg_ege_budget': df['avg_ege_budget'].median(),
    'cost_of_education_rub': df['cost_of_education_rub'].median(),
    'gdp_deflator': df['gdp_deflator'].median()
}, inplace=True)

```

```python
# список нужных столбцов
cols_to_convert = ['year', 'enrolled_paid', 'enrolled_budget']

# в числовой тип с обработкой ошибок
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# удаление строк с пропущенными значениями в этих столбцах
df = df.dropna(subset=cols_to_convert)

# Преобразуем к целому типу
df[cols_to_convert] = df[cols_to_convert].astype(int)


df.to_csv('./data/data_english_cleaned_1.csv', index=False)
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
# размер графиков по умолчанию
plt.rcParams['figure.figsize'] = (12, 6)

# общая численность студентов (платные + бюджетные)
df['total_enrolled'] = df['enrolled_paid'] + df['enrolled_budget']
students_by_year = df.groupby('year')['total_enrolled'].sum()

# общее количество студентов по всем университетам по годам'
plt.figure(figsize=(10, 6))
students_by_year.plot(kind='bar', color='skyblue')
plt.title('Общее количество студентов по всем университетам по годам')
plt.xlabel('Год')
plt.ylabel('Количество студентов')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2. средняя стоимость обучения по университетам.
# Группируем по университетам и считаем среднюю стоимость обучения
avg_cost_by_university = df.groupby('university')['cost_of_education_rub'].mean().sort_values(ascending=False)

# Оставим топ-20 университетов по стоимости
top_n = 20
top_avg_cost = avg_cost_by_university.head(top_n)

# горизонтальная диаграмма
plt.figure(figsize=(12, 8))
top_avg_cost.plot(kind='barh', color='coral')
plt.title(f'Средняя стоимость обучения (ТОП-{top_n} университетов)')
plt.xlabel('Стоимость обучения (руб.)')
plt.ylabel('Университет')
plt.gca().invert_yaxis()  # чтобы самый дорогой был сверху
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 3. динамика поступлений по годам
df_year_grouped = df.groupby('year')[['enrolled_budget', 'enrolled_paid']].sum()
df_year_grouped.plot(marker='o')
plt.title('Количество поступивших студентов по годам')
plt.ylabel('Количество студентов')
plt.xlabel('Год')
plt.grid(True)
plt.show()
```

```python
# 4. ТОП-10 университетов по среднему количеству студентов
top_universities = df.groupby('university')[['enrolled_budget', 'enrolled_paid']].mean().sum(axis=1).sort_values(ascending=False).head(10)
top_universities.plot(kind='bar')
plt.title('ТОП-10 университетов по среднему количеству студентов')
plt.ylabel('Среднее количество студентов')
plt.xlabel('Университет')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
# 5. Распределение стоимости обучения
sns.histplot(df['cost_of_education_rub'], bins=30, kde=True)
plt.title('Распределение стоимости обучения (руб)')
plt.xlabel('Стоимость обучения')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

```python
# 6. популярность направлений подготовки (program_group)
program_counts = df['program_group'].value_counts().head(15)
program_counts.plot(kind='bar')
plt.title('ТОП-15 направлений по количеству записей')
plt.ylabel('Количество')
plt.xlabel('Группа программ')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
